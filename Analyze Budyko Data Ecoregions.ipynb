### Front Matter

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
from scipy import stats
```

```python
data = pd.read_hdf('./budyko_working_er.hdf','df')
```

```python
# define a function for computing the EI from a given DI
def budyko(DI):
    return np.sqrt(DI*np.tanh(1/DI)*(1-np.exp(-DI)))
```

```python
data['PETpenvic_P'] = data['PETpenmanvic']/data['P'] # compute the dryness index
```

```python
data['ET_P'] = data['ET']/data['P'] # compute the evaporative index
```

## Plot the Budyko Curve

```python
DI = np.arange(0,50.001,0.001) # make a range to compute the budyko curve over
```

```python
EI = budyko(DI) # compute the Budyko Curve
```

## Compute Budyko anomaly

```python
data['pred_EI'] = data['PETpenvic_P'].apply(budyko) # compute the predicted evaportative index for each cell given its aridity
```

```python
data['Q_P'] = 1.-data['ET_P'] # compute actual runoff for each cell 
```

```python
data['Qpred_P'] = 1.-data['pred_EI'] # compute predicted runoff for each cell
```

```python
data['Qanom_P'] = data['Q_P']-data['Qpred_P'] # compute the anomoly (actual - predicted)
```

```python
data['Q'] = data.Q_P*data.P # compute Q not normalized by P
```

```python
data['Qanom'] = data.Qanom_P*data.P
```

### Fit a model to the streamflow anaomaly

```python
X = data.Sf # pull out the independant variable
y = data.Qanom_P # pull out the dependant variable

slope, intercept, r_value, p_value, std_err = stats.linregress(X,y) # fit the model

Sf = np.arange(0,1.001,.001) # generate a smooth snow fraction domain

mod = Sf*slope+intercept # generate predicted values
```

### Conclusion:

These figures suggest that snowmelt rate is key for predicting Qanomaly and that when precipitation is included in the calculation of water flux the relationship between water flux and Qanom falls apart. This then suggests that the magnitude and timing of peak SWE, which controls the snowmelt rate may have more bearing on whether a cell is under or over producing runoff than just the snowfall fraction of the cell. I think this points to more of a mechanistic understanding of the system where high snowmelt rates bring the soil to field capacity more frequenty and for longer, thus causing more infiltration and recharge of the shallow groundwater system. These cells then more efficiently generate runoff, which leads to positive streamflow anomalies in the Budyko framework.
