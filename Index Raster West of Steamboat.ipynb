### Problem:

If you take the unique lats and lons and make a big rectangle some parts of it will hang over the coastline. There will be more cells in the big rectangle than in the index...

### Solution

Fill the big rectangle with NaNs, fill only the cells that coorespond to the lat and lon variables with the index

```python
# Create the index matrix full of nans
indexer = np.zeros([len(np.unique(lat)),len(np.unique(lon))])
indexer[:] = np.NaN
```

```python
# create some indexer

lat_idx = np.arange(len(np.unique(lat)))
lon_idx = np.arange(len(np.unique(lon)))

unique_lats = np.unique(lat)
unique_lons = np.unique(lon)
```
