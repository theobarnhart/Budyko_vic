```python
# import global modules

import pandas as pd
import numpy as np
import glob

dates = pd.read_pickle('./timecode.pcl')
wyears = np.load('./wyears.npy')
```

```python
def newSf(fl):
    lat = float(fl.split('_')[-2]) # extract latitude from file name
    lon = float(fl.split('_')[-1]) # extract longitude from file name
    
    forcing_columns = ['P','Tmax','Tmin','W']
    
    dat = pd.read_table(fl, sep=' ', names=forcing_columns)
    
    dat['coef'] = (0.-dat.Tmin)/(np.abs(dat.Tmax-dat.Tmin))
    
    dat.loc[dat.coef<0,'coef'] = 0.
    dat.loc[dat.coef>1,'coef'] = 1.
    
    dat['snow'] = dat.P*dat.coef
    
    return lat,lon,dat.snow.sum()/dat.P.sum()
    
```

```python
# grab all the forcing cells:
pth = '/Users/barnhatb/research/vic/forcing/ascii/'

files = glob.glob(pth+'/*/Meteorology_*')
```

```python
res = view.map(newSf,files)
```
