```python
def getsoil(lat,lon):
    
    dat = soil.loc[(soil.lat == lat) & (soil.lon == lon),:]
    
    if dat.empty:
        infilt = np.NaN
        Ds = np.NaN
        Dsmax = np.NaN
        Ws = np.NaN
        c = np.NaN
        l1_expt = np.NaN
        l2_expt = np.NaN
        l3_expt = np.NaN
        l1_Ksat = np.NaN
        l2_Ksat = np.NaN
        l3_Ksat = np.NaN
        l1_phi_s = np.NaN
        l2_phi_s = np.NaN
        l3_phi_s = np.NaN
        l1_init_moist = np.NaN 
        l2_init_moist = np.NaN
        l3_init_moist = np.NaN
        elev = np.NaN
        l1_depth = np.NaN
        l2_depth = np.NaN
        l3_depth = np.NaN
        avg_T = np.NaN
        dp = np.NaN
        l1_bubble = np.NaN
        l2_bubble = np.NaN
        l3_bubble = np.NaN
        l1_quartz = np.NaN
        l2_quartz = np.NaN
        l3_quartz = np.NaN
        l1_bulkD = np.NaN
        l2_bulkD = np.NaN
        l3_bulkD = np.NaN
        l1_soil_density = np.NaN
        l2_soil_density = np.NaN
        l3_soil_density = np.NaN
        off_gmt = np.NaN
        l1_Wcr_FRACT = np.NaN
        l2_Wcr_FRACT = np.NaN
        l3_Wcr_FRACT = np.NaN
        l1_Wpwp_FRACT = np.NaN
        l2_Wpwp_FRACT = np.NaN
        l3_Wpwp_FRACT = np.NaN
        rough = np.NaN
        snow_rough = np.NaN
        annual_prec = np.NaN
        l1_resid_moisture = np.NaN
        l2_resid_moisture = np.NaN
        l3_resid_moisture = np.NaN
        fs_active = np.NaN
    else:
        infilt = float(dat.infilt)
        Ds = float(dat.Ds)
        Dsmax = float(dat.Dsmax)
        Ws = float(dat.Ws)
        c = float(dat.c)
        l1_expt = float(dat.l1_expt)
        l2_expt = float(dat.l2_expt)
        l3_expt = float(dat.l3_expt)
        l1_Ksat = float(dat.l1_Ksat)
        l2_Ksat = float(dat.l2_Ksat)
        l3_Ksat = float(dat.l3_Ksat)
        l1_phi_s = float(dat.l1_phi_s)
        l2_phi_s = float(dat.l2_phi_s)
        l3_phi_s = float(dat.l3_phi_s)
        l1_init_moist = float(dat.l1_init_moist) 
        l2_init_moist = float(dat.l2_init_moist)
        l3_init_moist = float(dat.l3_init_moist)
        elev = float(dat.elev)
        l1_depth = float(dat.l1_depth)
        l2_depth = float(dat.l2_depth)
        l3_depth = float(dat.l3_depth)
        avg_T = float(dat.avg_T)
        dp = float(dat.dp)
        l1_bubble = float(dat.l1_bubble)
        l2_bubble = float(dat.l2_bubble)
        l3_bubble = float(dat.l3_bubble)
        l1_quartz = float(dat.l1_quartz)
        l2_quartz = float(dat.l2_quartz)
        l3_quartz = float(dat.l3_quartz)
        l1_bulkD = float(dat.l1_bulkD)
        l2_bulkD = float(dat.l2_bulkD)
        l3_bulkD = float(dat.l3_bulkD)
        l1_soil_density = float(dat.l1_soil_density)
        l2_soil_density = float(dat.l2_soil_density)
        l3_soil_density = float(dat.l3_soil_density)
        off_gmt = float(dat.off_gmt)
        l1_Wcr_FRACT = float(dat.l1_Wcr_FRACT)
        l2_Wcr_FRACT = float(dat.l2_Wcr_FRACT)
        l3_Wcr_FRACT = float(dat.l3_Wcr_FRACT)
        l1_Wpwp_FRACT = float(dat.l1_Wpwp_FRACT)
        l2_Wpwp_FRACT = float(dat.l2_Wpwp_FRACT)
        l3_Wpwp_FRACT = float(dat.l3_Wpwp_FRACT)
        rough = float(dat.rough)
        snow_rough = float(dat.snow_rough)
        annual_prec = float(dat.annual_prec)
        l1_resid_moisture = float(dat.l1_resid_moisture)
        l2_resid_moisture = float(dat.l2_resid_moisture)
        l3_resid_moisture = float(dat.l3_resid_moisture)
        fs_active = float(dat.fs_active)
    
    return lat,lon,infilt,Ds,Dsmax,Ws,c,l1_expt,l2_expt,l3_expt,l1_Ksat,l2_Ksat,l3_Ksat,l1_phi_s,l2_phi_s,l3_phi_s,l1_init_moist,l2_init_moist,l3_init_moist,elev,l1_depth,l2_depth,l3_depth,avg_T,dp,l1_bubble,l2_bubble,l3_bubble,l1_quartz,l2_quartz,l3_quartz,l1_bulkD,l2_bulkD,l3_bulkD,l1_soil_density,l2_soil_density,l3_soil_density,off_gmt,l1_Wcr_FRACT,l2_Wcr_FRACT,l3_Wcr_FRACT,l1_Wpwp_FRACT,l2_Wpwp_FRACT,l3_Wpwp_FRACT,rough,snow_rough,annual_prec,l1_resid_moisture,l2_resid_moisture,l3_resid_moisture,fs_active
```
